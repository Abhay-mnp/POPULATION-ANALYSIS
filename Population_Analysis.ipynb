```python
highest_population = population_df.sort_values('Population (2020)', ascending=False).head(10)
highest_population
```

```python
highest_population_growth = population_df.sort_values('Net Change', ascending=False).head(10)
highest_population_growth
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# sns.set_style('darkgrid')
# matplotlib.rcParams['font.size'] = 14
# matplotlib.rcParams['figure.figsize'] = (9, 5)
# matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.figure(figsize=(20,10))
plt.title("Plot showing world's most populous countries, and their respective populations.")
sns.barplot(x = 'Country (or dependency)', y = 'Population (2020)', data = highest_population)
```

```python
plt.figure(figsize=(20,10))
plt.title("Scatter Plot with dots representing different countries of the world.")
sns.scatterplot(x='Land Area (Km²)', y='Population (2020)', data=population_df)
```

```python
high_population_density = population_df.sort_values('Density (P/Km²)', ascending=False).head(10)
high_population_density
```

```python
plt.figure(figsize=(20,10))
plt.title("Plot showing the world's most densely populated countries.")
sns.barplot(x = 'Country (or dependency)', y = 'Density (P/Km²)', data = high_population_density)
```

```python
data_of_india_bangladesh_pakistan = population_df.loc[(population_df['Country (or dependency)'] == 'India') |
                                              (population_df['Country (or dependency)'] == 'Bangladesh') |
                                              (population_df['Country (or dependency)'] == 'Pakistan')]
individual_land_area = data_of_india_bangladesh_pakistan[['Country (or dependency)', 'Land Area (Km²)']]
individual_land_area
```

```python
total_land_area = individual_land_area['Land Area (Km²)'].sum()
print(f"The total land area of present day India, Bangladesh and Pakistan is {total_land_area} Km².")
```

```python
individual_population = data_of_india_bangladesh_pakistan[['Country (or dependency)', 'Population (2020)']]
individual_population
```

```python
plt.pie(individual_population["Population (2020)"], labels = individual_population["Country (or dependency)"], autopct='%1.1f%%')
plt.title("Pie chart showing how population in the three countries India, Bangladesh, and Pakistan is distributed currently.")
plt.show()
```
